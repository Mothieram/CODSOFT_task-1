```python
X=titanic_data[['Pclass','Sex']]
Y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(random_state = 0)
log.fit(X_train, Y_train)
```

```python
import warnings
warnings.filterwarnings("ignore")


# Function to get user input and make a prediction
def predict_survival():
    pclass = int(input("Enter the passenger's class (1, 2, or 3): "))
    sex = int(input("Enter the passenger's gender (0 for female, 1 for male): "))

    # Create a feature vector with the user input
    user_input = [[pclass, sex]]

    # Make predictions using the model
    res = log.predict(user_input)

    if res == 0:
        print("Sorry, the passenger did not survive.")
    else:
        print("The passenger survived.")

predict_survival()
```
